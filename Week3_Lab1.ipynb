## 1. feature 2 개 (petal length, petal with) 를 이용한 퍼셉트론 모형 학습

```python
# 필요 패키지들 로딩하기 

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import Perceptron
```

```python
# feature 데이터에서 petal length 와 petal width 를 따로 추출하여 변수 X 에 저장하기

X=iris.data[:,(2,3)] 
```

```python
# target 값을 재코딩하기 (setosa =1, 나머지 =0)

y=(iris.target==0)
y=y.astype(np.int)
print(y)
```

```python
# X 의 분포를 표현하기

plt.scatter(X[:,0],X[:,1])
```

```python
# target 값에 따른 X 의 분포를 표현하기

plt.scatter(X[:,0][y==0],X[:,1][y==0],c='blue')
plt.scatter(X[:,0][y==1],X[:,1][y==1],c='red')
plt.xlabel("$X_1$") # petal length
plt.ylabel("$X_2$") # petal width
plt.plot()
```

```python
#모형 지정하기
    
per_clf=Perceptron()
```

```python
#모형 학습하기

per_clf.fit(X,y)
```

## 2. 학습 결과 확인하기

```python
# 학습된 가중치와 편향값 확인하기 

print(per_clf.coef_)
print(per_clf.intercept_)
```

```python
# 위 그림을 그대로 재현하되, 결정 경계식 (decision boundary) 의 수식을 추가하기 

xx, yy = np.meshgrid(np.arange(0, 8, 0.01), np.arange(0, 3.0, 0.01))

# 이 구간에서 
Z = per_clf.predict(np.c_[xx.ravel(), yy.ravel()]) #np.c_는 가로로 합치는것. 
Z = Z.reshape(xx.shape)

plt.figure(1)

# 바운더리 출력
plt.pcolormesh(xx, yy, Z, cmap = "RdBu", vmin = -.2, vmax = 1.2)

# 데이터 점 출력
plt.scatter(X[:,0], X[:, 1], c=y, s=50, 
            cmap = "RdBu", vmin = -.2, vmax = 1.2,
            edgecolor="white", linewidth=1)

# 축 설명 (꽃 잎)
plt.xlabel("petal length")
plt.ylabel("petal width")

# 결정경계 식 추가 

plt.title('Decision boundary: $2-0.7x_1-1.2x_2=0$')
```

```python
# training error 출력하기 

np.sum(per_clf.predict(X) != y)
```

```python
# petal length =2 cm, petal with=0.5 cm 일때의 iris 종류 예측하기

y_pred=per_clf.predict([[1, 0.2]])
y_pred
```
